```python
df=df.rename(columns={"students_id":"ID","name":"Name","age":"Age","gender":"Gender","grade":"Class","math_score":"Math_Score","english_score":"English_Score","science_score":"Science_Score","enrolled_date":"Enrolled_Date","remarks":"Remarks"})
df
```

```python
df["Gender"]=df["Gender"].replace({"male":"M","female":"F","MALE":"M","FEMALE":"F","Male":"M","Female":"F"})
df
```

```python
df["Class"]=df["Class"].replace({"10th":"X","11th":"XI","12th":"XII","10":"X","11":"XI","12":"XII"})
df
```

```python
df["Age"]=df["Age"].fillna(0).astype(int)
df
```

```python
df["Math_Score"]=df["Math_Score"].fillna(0).astype(int)
df
```

```python
df["English_Score"]=df["English_Score"].replace({"missing":"0"})
df
```

```python
df["English_Score"]=df["English_Score"].fillna(0).astype(int)
print(df.head())
```

```python
df["Name"]=df["Name"].str.strip()
df
```

```python
df[["First_Name","Last_Name"]]=df["Name"].str.split(" ",n=1,expand=True)
df.drop(columns=["Name"],inplace=True)
df
```

```python
df["Remarks"]=df["Remarks"].replace({"excellent":"A","good":"B","GOOD":"B","average":"C","Average":"C","poor":"D","needs improvement":"E","good student":"O"})
df
```

```python
df["Enrolled_Date"]=df["Enrolled_Date"].astype(str).str.replace("/","-",regex=True)
df
```

```python
df["Last_Name"]=df["Last_Name"].str.strip(".")
df
```

```python
df=df[
  [
        "student_id",
        "First_Name",
        "Last_Name",
        "Age",
        "Gender",
        "Class",
        "Math_Score",
        "English_Score",
        "Science_Score",
        "Enrolled_Date",
        "Remarks"
  ]
]
df
```

```python
df["Enrolled_Date"]=pd.to_datetime(df["Enrolled_Date"],dayfirst=True,errors='coerce').dt.strftime('%d-%m-%y').fillna(df['Enrolled_Date'])
df
```
